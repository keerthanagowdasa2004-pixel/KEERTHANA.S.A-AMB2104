```python
import pandas as pd
data={
    'Farm':['Farm A','Farm B','Farm C','Farm D'],
    'Crop':['Wheat','Rice','Wheat','Maize'],
    'Area_hectares':[10,15,8,12],
    'Production_tons':[30,60,24,36]
}
df=pd.DataFrame(data)
print(df)
df.head()
```

```python
df.head(2)
```

```python
df.tail(2)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['yield_ton_per_hectare']=df['Production_tons']/df['Area_hectares']
df
```

```python
df.iloc[[1,3],[0,1,2]]
```

```python
df.iloc[[1,3]]
```

```python
#filter data
#farms growing wheat
print(df['Crop']=='Wheat')
print()
wheat_farms=df[df['Crop']=='Wheat']
print(wheat_farms)
```

```python
high_yield=df[df['Yield_ton_per_hectare']>3]
high_yield
```

```python
sorted_df=df.sort_values(by='Production_tons',ascending=False) #sort farms by highest production
sorted_df
```

```python
#GroupBy operation-avg yield per crop
avg_yield=df.groupby('Crop')['Yield_ton_per_hectare'].sum()
avg_yield
```
