```python
#define a function
def max(num1,num2):
    if num1>num2:
        result=num1
    else:
        result=num2
    return result
z=max(22,44) #invoke a function
print(z)
```

```python
#with parameter without return type
def fact(m):
    fact=1
    if (m>0):
        for i in range(1,m+1):
           fact=fact*i
        print("factorial of", m ,"is",fact)
    else:
         print('enter a valid number')
n=int(input('enter a number'))
fact(n)
```

```python
#with parameter with return type
def fact(m):
    fact=1
    if (m>0):
        for i in range(1,m+1):
           fact=fact*i
        return fact #print
    else:
         print('enter a valid number')
n=int(input('enter a number'))
z=fact(n) #store the returned value in z
print(z) #print the value of z
```

```python
#without parameter without return type
def fact():
    m=int(input('enter a number'))
    fact=1
    if (m>0):
        for i in range(1,m+1):
           fact=fact*i
        print("factorial of", m ,"is",fact)
    else:
         print('enter a valid number')

fact()

```

```python
#without parameter with return type
def fact():
    m=int(input('enter a number'))
    fact=1
    if (m>0):
        for i in range(1,m+1):
           fact=fact*i
        return fact
    else:
         print('enter a valid number')

z=fact()
print(z)
```

```python
#scope of variables  #local variable
def func(x):
    print('Local x is :',x)
    x=2
    print('changed local x to:',x)
x=50
func(x)
print('x is still:',x)
```

```python
#global variable 
def func1():
    global y
    print('Y is:',y)
    y=2
    print('changed y to:',y)
y=50
func1()
print('value of y is:',y)
```

```python
total=0
def sum(arg1,arg2):
    total=arg1+arg2
    print('inside the function is total:',total)
    return(total)
sum(10,20)
print('outside the function global total:',total)
```

```python
#types of function or formal arguments(parameter)
#(1.required arguments) 
def printme(str):
    print(str)
    return
#printme() #statement gives out the error
printme('hello')
printme(1234)
printme('hello','hi') #gives error as we have only one argument
```

```python
#keyword or named arguments
def printme(str):
    print(str)
    return
printme('hello')
printme(str='my string')
printme(abc='my string')#error because abc is unknown argument
```

```python
#default argument ---ALWAYS AT END 
def func(a,b=5,c=20):
    print('a is :',a,'b is :',b,'c is :',c)
func(3,7)
func(25,c=24) 
func(c=50,a=20)
```

```python
#variable length argument
def check(name,*marks):# '*' is used to hold multiple values
    print(name)
    print(marks)
check('xyz',10,20,30)
```
