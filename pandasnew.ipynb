```python
import pandas as pd
df=pd.read_csv('agri_data.csv')

#custom styled bar plot
import matplotlib.pyplot as plt

yearly_production = df.groupby("Year")["Production_Tons"].sum()

plt.figure(figsize=(5,3))
bars = plt.bar(yearly_production.index,
               yearly_production.values,
               color="lightgreen",
               edgecolor="black")

#add value labels on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height,
             str(height),
             ha="center",
             va="bottom")

plt.xlabel("Year")
plt.ylabel("Production(tons)")
plt.title("yearly production")
plt.show()
```

```python
import pandas as pd
import numpy as np
#create data with missing values
data=pd.DataFrame({
    'plot':['A','B','C','D','E'],
    'Yield_kg':[2500,np.nan,2800,np.nan,2600],#nan means not a number
    'rainfall':[150,200,np.nan,180,120],
    'temperature':[28,30,29,np.nan,27]
})
print('data with missing values:')
print(data)
    
```

```python
#detecting missing data
print('\n===detecting missing data===')
print('is null?')
print(data.isnull())
print(f'\n{data.isnull().sum()}')
```

```python
#fillna = to fill wherever nan is there
#fill with mean
print('\n===Filling Stratergies===')
data_mean=data.copy()
print(data_mean['Yield_kg']) 
data_mean['Yield_kg']=data_mean['Yield_kg'].fillna(data_mean['Yield_kg'].mean())
print('\n','Fill with mean')
print(data_mean['Yield_kg'])

```

```python
#filled with median
data_median=data.copy()
print(data_median['rainfall'])
data_median['rainfall']=data_median['rainfall'].fillna(data_median['rainfall'].median())
print(data_median['rainfall'])
print('\n','fill with median')
print(data_median)
      
```

```python
#forwardfill(use previous value)
data_ffill=data.copy()
print(data_ffill)
data_ffill=data_ffill.fillna(method='ffill')
print('\nForward fill:')
print(data_ffill)
#ffill-it replaces missing value(NaN) with the last valid(non-missing)value above it in the column
    
```

```python
#dropping missing data
print('\n===Dropping missing rows===')
data_dropped=data.dropna()
print(data_dropped)
print(f'original rows:{len(data)},after drop:{len(data_dropped)}')
```
