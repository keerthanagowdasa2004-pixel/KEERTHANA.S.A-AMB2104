```python
clean_codes=[]
for code in codes:
    if code.startswith('CRP_'):
        clean_codes.append(code[4:])
print(clean_codes)
```

```python
import random
#initialize inventory and money
inventory={}
total_money=0
#fish types and prices
fish_prices={
    'Rohu':12,
    'Catla':18,
    'Mrigal':25
}
def showmenu():
    print('\n Fishing game menu')
    print('1.Catch fish')
    print('2.View Inventory')
    print('3.Sell fish')
    print('4.Exit')
def catchfish():
    fish=random.choice(list(fish_prices.keys()))
    inventory[fish]=inventory.get(fish,0)+1
    print(f"You caught a {fish}!")
def your_inventory():
    if inventory=={}:
        print('No fish')
    else:
        print('Your inventory')
        for fish,count in inventory.items():
            print(f"{fish}:{count}")
def sell_fish():
    global total_money
    if inventory=={}:
        print('No fish to sell')
    else:
        earnings=0
        for fish,count in inventory.items():
            earnings+=fish_prices[fish]*count
        total_money+=earnings
        inventory.clear()
        print(f"you sold all fish for ${earnings}!")
def end_game():
        print(f"Final money earned:${total_money}")
        print("Thanks for playing!")
    

    


showmenu()
#game loop
while True:
    choice=int(input('Choose an option'))
    if choice==1:
        catchfish()
       
    elif choice==2:
          your_inventory()
        
        
        
    elif choice==3:
          sell_fish()
    elif choice==4:
           end_game()
           break
    else:
        print("Invalid choice.Try again.")
                
    
```
