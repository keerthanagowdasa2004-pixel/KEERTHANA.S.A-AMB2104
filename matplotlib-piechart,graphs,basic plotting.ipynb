```python
#piechart
import matplotlib.pyplot as plt
labels=['Rice','Wheat','Maize','Jowar','Cotton']
sizes=[33,25,52,12,48]
seperated=(0,0,0,0,0.1)
plt.pie(sizes,labels=labels,autopct='%1.1f%%',explode=seperated)
plt.show()
```

```python
#piechart
import matplotlib.pyplot as plt
slices=[7,2,1,13]
activities=['weeding','ploughing','earthing up','irrigation']
cols=['r','m','c','b']
plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow=True,
        explode=(0,0.1,0,0),
        autopct='%1.1f%%')
plt.title('This is a pie chart')
plt.show()
```

```python
import numpy as np
years=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
open_sensex=[17473,20622,15535,19513,21222,27486,26102,26711,34090,36162,41349]
close_sensex=[20509,12990,84300,23893,38921,33993,19304,23565,48022,30022,59302] 
ind=np.arange(len(years))
print(ind)
width=0.3
plt.bar(ind,open_sensex,width,label='net')
plt.bar(ind+width,close_sensex,width,label='gross',color='g')
plt.title('Production per year')
plt.ylabel('production')
plt.xlabel('year')
plt.xticks(ind,years,rotation=70)
plt.legend()
plt.show()
```

```python
import numpy as np
years=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
open_sensex=[17473,20622,15535,19513,21222,27486,26102,26711,34090,36162,41349]
close_sensex=[20509,12990,84300,23893,38921,33993,19304,23565,48022,30022,59302] 
ind=np.arange(len(years))
print(ind)
width=0.3
plt.barh(ind,open_sensex,width,label='net')
plt.barh(ind+width,close_sensex,width,label='gross',color='g')
plt.title('Production per year')
plt.ylabel('production')
plt.xlabel('year')
plt.xticks(ind,years)
plt.legend()
plt.show()
```
