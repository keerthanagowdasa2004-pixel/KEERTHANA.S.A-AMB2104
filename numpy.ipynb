```python
#elementwise arithmetic operations
import numpy as np
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
print(arr1)
print(arr2)
print(arr1+arr2) 
print(arr1-arr2) 
print(arr1*arr2) 
print(arr1/arr2)
print(arr1@arr2) #actual matrix multiplication
concatenated_arr=np.concatenate((arr1,arr2),axis=0)
#concatenate along rows
print(concatenated_arr)
print()
concatenated_arr=np.concatenate((arr1,arr2),axis=1)
#concatenate along columns
print(concatenated_arr)
print()
arr=np.array([[1,2,3],[4,5,6]])
transposed_arr=arr.transpose() #interchange row elements to column elements and vice-versa
print(transposed_arr)
print()
flattened_arr=arr.flatten() # converts multidimension to single dimension
print(flattened_arr)
print()


```

```python
#crop yields for 4 farms over 5 seasons(in tons)
import numpy as np
crop_yields=np.array([
    [12.5,13.4,14.4,13.2,15.5],
    [10.1,11.2,10.3,12.4,11.9],
    [15.2,13.4,15.6,16.3,15.1],
    [11.1,10.2,12.1,13.2,11.1]
])
print('Crop yields array(tons):',crop_yields)
print(f'\nShape:{crop_yields.shape}')
print(f'Datatype:{crop_yields.dtype}')
print(f'Total elements:{crop_yields.size}')
print(f'Number of dimensions:{crop_yields.ndim}')
    

    
```

```python
import numpy as np
total_yield=np.sum(crop_yields)
avg_yield_per_farm=np.mean(crop_yields,axis=1) #row
avg_yield_per_season=np.mean(crop_yields,axis=0) #column
max_yield=np.max(crop_yields)
min_yield=np.min(crop_yields)

print(f'\nTotal production:{total_yield}tons')
print(f'Average per farm:{avg_yield_per_farm}tons')
print(f'Average per season:{avg_yield_per_season}tons') 
print(f'Maximum yield:{max_yield}tons')
print(f'Minimum yield:{min_yield}tons')
```

```python
best_farm=np.argmax(avg_yield_per_farm) #argmax() does not return the max value itself-it returns where that value is located
worst_farm=np.argmin(avg_yield_per_farm)
print('best farm:',best_farm)
print('worst farm:',worst_farm)
print('\nbest performing farm:farm',best_farm +1)
```

```python
#boolean indexing:find exceptional yields
exceptional=crop_yields>15
print(exceptional)
print('\nExceptional yields(>15):')
print(crop_yields[exceptional])
```

```python
#conditional operations:classify yields
categories=np.where(crop_yields>16,'high',
                    np.where(crop_yields>14,'medium','low'))
print('\nyield_categories:')
print(categories)
```
